```python
import numpy as np
import holoviews as hv
from Julia_funcs import lambda_par
from numba import njit
hv.extension('bokeh')
```

```python
w, h = 800, 800 #   
maxit = 5000 #  
c = complex(-1,-0.1) #    
x_min, y_min = -2, -2 #    
x_max, y_max = 2, 2 # 
```

```python
x_range = (x_min, x_max)
y_range = (y_min, y_max)
```

```python
def fractal(x_range: float, y_range: float, maxit: int, c: complex):
    '''
    :param x_range:  xmin  xmax
    :param y_range:  ymin  ymax
    :param it: int, - 
    :return:
     
    '''
    frac = lambda_par(x_range, y_range, w, h, maxit, c)
    return hv.Image(frac, bounds=(x_range[0], y_range[0], x_range[1], y_range[1]))
```

```python
def create_dynamic_map(x_range: float, y_range: float, maxit: int, c: complex):
    '''
    Create DynamicMap
    
    :param x_range:  xmin  xmax
    :param y_range:  ymin  ymax
    :param it: int, - 
    :return:  
    '''
    range_xy = hv.streams.RangeXY(x_range=x_range, y_range=y_range)
    dmap = hv.DynamicMap(lambda x_range, y_range, maxit=maxit, c=c: fractal(x_range, y_range, maxit, c),
                          streams=[range_xy])
    return dmap
```

```python
create_dynamic_map(x_range, y_range, maxit, c).opts(hv.opts.Image(width=w, height=h, cmap='twilight', logz=True))
```
